IMPORTING THE DEPENDENCIES


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND PROCESSING


```python
 #LOADING THE CSV DATA TO A PANDAS DATAFRAME
 heart_data=pd.read_csv('/Heart disease prediction.csv')
```

```python
#printing first 5 rows of the dataset
heart_data.head()
```

```python
#printing last 5 rows of the dataset
heart_data.tail()
```

```python
#number of rows and columns in the dataset
heart_data.shape
```

```python
#getting info about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measures of the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts()
```

1-->Defective heart
0-->Healthy heart


Splitting the Features and Target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
print(X)
```

```python
print(Y)
```

SPLITTING THE DATA INTO TRAINING DATA AND TEST DATA


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING

LOGISTIC REGRESSION

```python
model=LogisticRegression()
```

```python
#training the LogisticRegression model with training data
model.fit(X_train,Y_train)
```

ACCURACY SCORE


MODEL EVALUATION


```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test data:',test_data_accuracy)
```

BUILDING A PREDICTIVE SYSTEM

```python
input_data=(58,0,0,100,248,0,0,122,0,1,1,0,2)

#change the input data to a numpy array for manipulations
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('This Person is predicted to not have a Heart Disease')
else:
  print('This Person is predicted to have a Heart Disease')
```
